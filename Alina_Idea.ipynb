**UniTradeGPT**

Every college campus has the same problem — cluttered Facebook groups, messy spreadsheets, and scam-filled chats where students try to sell or trade things like furniture, tech, textbooks, or even lease transfers. It’s chaotic, unreliable, and impossible to keep organized.

UniTradeGPT changes that. It’s a secure, student-only marketplace built specifically for college communities. Whether you’re buying a used laptop, reselling a dorm couch, finding a ride, or swapping subleases, everything happens in one trusted, easy-to-use platform.

Listings are verified through student emails, the interface is sleek and intuitive, and smart filters make finding what you need effortless. By cutting out bots and scammers and bringing all campus exchanges into one space, UniTradeGPT keeps your college economy safe, fast, and local.

In short: we’re making student-to-student selling simple again — because your time on campus shouldn’t be wasted scrolling through chaos.
