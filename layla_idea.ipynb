DocuAssist is a document-managing chatbot for researchers. Imagine you're a PhD student who has spent months reviewing literature for your dissertation. You remember reading something you now want to cite, but you don't remember which of the dozens of papers it came from. Instead of pouring hours back into re-reading those papers, DocuAssist can retrieve the information and where it came from for you in seconds. Or imagine you're a lawyer new to a firm that frequently references the same set of cases. Instead of digging through folders or scouring hundreds of PDFs or case files, you can simply ask DocuAssist questions in natural language to retrieve information from your files. DocuAssist will output only what exists in the corpus you build, and let you know exactly which file the information came from.
