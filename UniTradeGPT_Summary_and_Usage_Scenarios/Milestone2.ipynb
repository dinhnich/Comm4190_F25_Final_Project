**College Marketplace - UniTradeGPT**

**Pitch:** Every college campus has the same problems when it comes to buying and selling items. There are so many cluttered groupchats, scam-filled chats, and disorganized servers where students try to sell or trade things like furniture, technology, textbooks, clothes, or even lease transfers. It’s chaotic, unreliable, and impossible to navigate and keep organized.

UniTradeGPT changes that. It’s a secure, student-only marketplace built specifically for college communities. Whether you’re buying a used laptop, reselling a dorm couch, finding a ride, or swapping subleases, everything happens in one trusted, easy-to-use platform.
Listings are verified through student emails, the interface is sleek and intuitive, and smart filters make finding what you need effortless. By acting as a middleman, assisting users, and bringing all campus exchanges into one space, UniTradeGPT keeps your college economy safe, fast, and local.

In short: we’re making student-to-student selling simple again — because your time on campus shouldn’t be wasted scrolling through chaos.

**Description of the target users**: The main demographic of the UniTradeGPT users will be college students. Shared marketplaces would either be college/university students all from the same institution or regionally connected institutions. Students would include both undergraduate and graduate students in order to increase the pool of both buyers and sellers. The goal of grouping students by institution or region is to ensure transactions happen locally for the convenience of all parties involved. The majority of users would be those looking to buy second-hand items to minimize costs of a somewhat short-term living arrangement (1-5 years) and those looking to sell items in too good condition to just throw away. 

**More detailed summary of what the application will do**: UniTradeGPT is a secure, student-exclusive marketplace where users register using their verified student email. The platform is organized into regional communities based on cities or areas, allowing students from nearby schools to connect and trade within the same network. For instance, students from Penn and Drexel would share the same Philadelphia community. Within the app, students can list items for sale, trade, or lease, and browse listings from other verified students.
In addition to the marketplace functions, UniTradeGPT integrates an LLM that enhances user interactions. The LLM serves as a smart assistant, helping to manage communication and logistics between buyers, sellers, and traders. It can schedule meetups or deliveries, answer common questions, generate simple contracts, filter and compare items, collect user feedback, and even provide interior design suggestions. This creates a streamlined and low-hassle trading experience for students.

**What data (from users, web, other sources, etc) will be needed**
Data from Users
Verification data: Student email (to confirm enrollment and assign community), phone number, and name.
Profile data: Associated school/university, city or area, and distance radius for searching nearby listings.
Marketplace data: Item listings created by users including:
Title, description, category (e.g., furniture, electronics)
Price, quality/condition, weight, dimensions, photos
Listing status (active/sold/traded) and timestamps
Interaction data:
Messages between buyers/sellers/traders (facilitated by the LLM)
Ratings, reviews, and feedback on completed transactions
Search history, saved items, and preferences (if user opts in)


Logistics data: Preferred times and locations for pickup/drop-off (used by the LLM for scheduling).
Data from the Web or External Sources
Address and location data: For distance and proximity calculations between users.
Geocoding and mapping data: APIs or datasets to estimate distances and routes.
Reference knowledge:
Common room setup data (e.g., what’s typically needed in dorms, kitchens, or bedrooms).
General marketplace and pricing trends to help the LLM compare or suggest fair values.
Furniture and product taxonomies, color/material standards, and unit conversion data.
Data from Other System Components
Analytics and safety systems: Collect aggregate usage statistics, detect suspicious activity, and enforce verified-student-only policies.

**What is the role of LLMs**
The role of the LLM will be to retrieve information from the platform to assist the user with their request. The LLM will have access to user information, including profile data, item listings, and memory of interactions; once the user inputs their request, the LLM’s capabilities can include suggesting item listings, evaluating listings (including providing cost-benefit analysis), drafting between-user agreements, planning pick up/drop off logistics, coordinating schedules, and more. 

**How your product will differ from both existing non-LLM powered systems in the domain and from just using a general purpose LLM for the task (e.g. Gemini, Claude or ChatGPT)**
Most college students who are looking to buy, sell, or lease utilize social media like GroupMe or Facebook to do so. These group chats become cluttered quickly, and listings can sometimes be drowned out by new messages on the same day that they were posted. Additionally, students who make use of these chats must navigate scam listings, and even if they are successfully avoided, the presence of scams on these apps can lead to general distrust of using them. Apps that are solely dedicated to listing items (e.g. BuyNothing) similarly have scam issues associated with them, and navigating listings or interacting with buyers/sellers could instead be heavily streamlined with an LLM-powered system.
Lastly, general purpose LLMs do not have access to the aforementioned listings or item requests on those existing applications, and users of general purpose LLMs cannot make item listings on them or interact with other users data; thus, neither existing non-LLM powered applications nor general purpose LLMs are able to assist users in the way UniTradeGPT will offer, which is providing trustworthy and efficient assistance in a student’s search to buy or sell.
